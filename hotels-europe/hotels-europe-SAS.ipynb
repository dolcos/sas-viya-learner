```sas
%let features_csv = hotels-europe_features.csv; 
%let price_csv = hotels-europe_price.csv;

%put &=features_csv;
%put &=price_csv;

proc import datafile="&INDIR./&features_csv."
    out=WORK.features
    dbms=csv
    replace;
    guessingrows=max;
    getnames=yes;
run;

proc import datafile="&INDIR./&price_csv."
    out=WORK.price
    dbms=csv
    replace;
    guessingrows=max;
    getnames=yes;
run;
```

```sas
title "features (first 5 rows)";
proc print data=work.features (obs=5); run;

title "Contents price";
proc contents data=work.price; run;

title "Contents features";
proc contents data=work.features; run;
```

```sas
proc sql;
    create table work.hotels_europe as
    select a.*, b.*
    from work.price    as a          /* a.hotel_id is numeric */
    left join work.features as b     /* b.hotel_id is character */
      on a.hotel_id = input(b.hotel_id, best32.);
quit;

/* remove the duplicate column from features_df (the 2nd one) */
data work.hotels_europe;
    set work.hotels_europe;
    drop hotel_id1;   /* SAS will usually rename the duplicate to hotel_id1 */
run;

title "hotels_europe (first 5 rows)";
proc print data=work.hotels_europe (obs=5); run;
```

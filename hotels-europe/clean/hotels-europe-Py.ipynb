```python
# --- import libraries ---
import pandas as pd

# --- read CSVs ---
features_df = pd.read_csv("hotels-europe_features.csv")
price_df    = pd.read_csv("hotels-europe_price.csv")

# --- preview first rows (like head()) ---
print("Features:")
print(features_df.head(), "\n")

print("Price:")
print(price_df.head(), "\n")

# --- left join on hotel_id ---
hotels_europe = price_df.merge(features_df, on="hotel_id", how="left")

# --- inspect structure (like glimpse()) ---
print(hotels_europe.info())

# --- optional: preview combined data ---
print("\nFirst 5 joined rows:")
print(hotels_europe.head())
```

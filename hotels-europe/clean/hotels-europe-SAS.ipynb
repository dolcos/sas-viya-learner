```sas
/* ---- 0) Paths (edit if your CSVs live elsewhere) ---- */ 
%let features_csv = hotels-europe_features.csv; 
%let price_csv = hotels-europe_price.csv;

%put &=features_csv;
%put &=price_csv;

/* ---- 1) Read CSVs into WORK. tables ---- */
proc import datafile="&features_csv."
    out=work.features_df
    dbms=csv
    replace;
    guessingrows=max;
    getnames=yes;
run;

proc import datafile="&price_csv."
    out=work.price_df
    dbms=csv
    replace;
    guessingrows=max;
    getnames=yes;
run;

/* ---- 2) Peek at the data (like head()) ---- */
title "features_df (first 5 rows)";
proc print data=work.features_df (obs=5); run;
```

```sas
proc sql;
    create table work.hotels_europe as
    select a.*, b.*
    from work.price_df as a
    left join work.features_df as b
        on a.hotel_id = b.hotel_id;
quit;

/* remove the duplicate column from features_df (the 2nd one) */
data work.hotels_europe;
    set work.hotels_europe;
    drop hotel_id1;   /* SAS will usually rename the duplicate to hotel_id1 */
run;
```
